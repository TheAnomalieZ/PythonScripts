```python
directoryPath="/home/mani/JFR_CSV/cep/newpausetimes/"
list =[]
i=0
for file_name in glob.glob(directoryPath+'*.csv'):
    f=open(file_name, 'r')
    states = []
    try:
        reader = csv.reader(f)
        for row in reader:
            states.append(row)
    finally:
        f.close()
    list.insert(i,np.array(states,dtype='float32'))
    i+=1
```
