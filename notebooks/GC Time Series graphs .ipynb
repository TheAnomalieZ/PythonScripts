```python
import numpy as np
import matplotlib.pyplot as plt
import math as mm
import csv
%matplotlib notebook
```

```python
def openFile(file):
    f = open(file)
    try:
        reader = csv.reader(f)
        floats = []
        for row in reader:
            floats.append(row)
    finally:
        f.close()
    return floats
```

```python
list =[]
# for j in range(1,8):
#     name = 'cep/cep00'+str(j)+'_pausetime.csv'
#     print (name)
#     list.insert(j,name)
    
list.insert(1,'anomalyGC_series.csv')    
```

```python
series = []
for i in range(0,len(list)):
    series.insert(i,np.array(openFile('../data/'+list[i])))
    print(series[0].shape)
    
    x = np.arange(1, (series[i].size+1)/2, 1)
    print(x.shape)

    y = series[i][:,0]
    print(y.shape)
    z= series[i][:,1]
    plt.plot(x,y,'r')
    plt.plot(x,z,'g')
    # plt.plot(numberoftest,-80,'g')
    # plt.xlim(2000,10000, 100)
    plt.xlabel('Time')
    plt.ylabel('PauseTime')
    plt.title(list[i]+' Pause Time series')
    plt.show()

 
```
